```python
def greet_user(name="user"):
    print(f"Welcome, {name}!")

greet_user()
greet_user("Jyotsna")
```

```python
def print_names(*names):
    print("Names passed:", names)
    for name in names:
        print("Hello", name)

print_names("Jyotsna", "bvrit", "python")
```

```python
def print_info(**info):
    print("User Info:")
    for key, value in info.items():
        print(f"{key} = {value}")
    print(info)

print_info(name="Jyotsna ", age=18 )
```

```python
def show_details(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

show_details(1, 2,'A', 3.14, name="jYOTSNA", city="hyderabad")
```

```python
def add(a, b):
    return a + b

def doubling(n):
    return add(n, n)

print("Sadding 2 with itself:", doubling(2))
```
