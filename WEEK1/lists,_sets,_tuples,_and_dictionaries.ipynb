```python
fruits = ['apple', 'banana', 'cherry', 'mango']
print("Original list:", fruits)

print("First fruit:", fruits[0])
print("third fruit:",fruits[2])
```

```python
fruits.append('papaya')
print("After appending:", fruits)
```

```python
fruits.remove('banana')
print("After removing banana:", fruits)

print("Number of fruits:", len(fruits))
```

```python
measurements = (5,10,15)
print("Tuple:", measurements)
```

```python
print("Second element in tuple:",measurements[1])
```

```python
unique_numbers = {1, 2, 3, 2, 1,7}
print("Set (no duplicates):", unique_numbers)

```

```python
unique_numbers.add(4)
print("After adding 4:", unique_numbers)
unique_numbers.add(9)
print("after adding 9:",unique_numbers)
```

```python
unique_numbers.remove(7)
print("After removing 7:", unique_numbers)
```

```python
set_a = {1, 2, 3}
set_b = {3, 4, 5}
print("Union:", set_a | set_b)
print("Intersection:", set_a & set_b)
print("Difference (A - B):", set_a - set_b)
```

```python
student = {
    'name': 'jyotsna',
    'age': 18,
    'course': 'Python'
}
print("Dictionary:", student)
```

```python
student['grade'] = 'A'
print("After adding grade:", student)
del student['age']
print("After deleting age:", student)

for key, value in student.items():
    print(f"{key}: {value}")
```
